## Задача 2:

Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions

```python
expenses = [100, 125, -90, 345, 655, -1, 0, 200]
sum_of_expenses = [i if i > 0 else 0 for i in expenses]
print('Отредактированный массив:', sum_of_expenses)
print('Общая сумма затрат:', sum(sum_of_expenses))
```

Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.

```python


def monthly_income(date: list, income: list) -> dict[str, int]:
    result = {}
    for d, i in zip(date, income):
        if d[5:7] in result.keys():
            result[d[5:7]] = result[d[5:7]] + i
        else:
            result[d[5:7]] = i
    return result

months = {'01': 'Январь', '02': 'Февраль', '3': 'Март', '4': 'Апрель', '5': 'Май',
        '6': 'Июнь', '7': 'Июль', '8': 'Август', '09': 'Сентябрь', '10': 'Октябрь',
        '11': 'Ноябрь', '12': 'Декабрь'}

my_dict = (monthly_income(date, income))

new_dict = {}

for key, value in my_dict.items():
    month_name = months.pop(key)
    new_dict[month_name] = value
    
print('Выручка по месяцам:', new_dict)

```
